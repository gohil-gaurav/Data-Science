Descriptive statistics helps us summarize and understand the main features of a dataset.

It provides simple summaries about the data such as average, spread, and distribution.

---

## Topics Covered

1. Mean
2. Median
3. Mode
4. Variance
5. Standard Deviation
6. Data Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Creating Sample Dataset

We will create a dataset to demonstrate statistical concepts.

```python
data = [10, 20, 20, 30, 40, 50, 60, 70, 80, 100]

df = pd.DataFrame(data, columns=["Values"])

df
```

## 2. Mean (Average)

The mean is the average of all values in the dataset.

```python
mean_value = df["Values"].mean()

print("Mean:", mean_value)
```

## 3. Median

The median is the middle value when data is sorted.

```python
median_value = df["Values"].median()

print("Median:", median_value)
```

## 4. Mode

The mode is the most frequently occurring value in the dataset.

```python
mode_value = df["Values"].mode()

print("Mode:")
print(mode_value)
```

## 5. Variance

Variance measures how spread out the data is from the mean.

```python
variance_value = df["Values"].var()

print("Variance:", variance_value)
```

## 6. Standard Deviation

Standard deviation shows how much variation exists in the dataset.

```python
std_value = df["Values"].std()

print("Standard Deviation:", std_value)
```

## 7. Summary Statistics

Pandas provides a built-in method to get all statistics at once.

```python
df.describe()
```

## 8. Data Distribution Visualization

```python
plt.hist(df["Values"])
plt.title("Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

## 9. Box Plot

Box plots help identify spread and outliers in the data.

```python
sns.boxplot(x=df["Values"])
plt.title("Box Plot")
plt.show()
```

## Conclusion

In this notebook we learned:

- Mean (average value)
- Median (middle value)
- Mode (most frequent value)
- Variance (spread of data)
- Standard deviation (data variation)
- Data distribution using plots

These statistical measures are essential for understanding datasets before applying machine learning models.
