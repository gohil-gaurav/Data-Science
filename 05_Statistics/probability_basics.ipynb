```python
import numpy as np
```

## 1. Basic Probability

Probability is defined as:

Probability = (Number of favorable outcomes) / (Total outcomes)

```python
# Example: Probability of getting a head in a coin toss

total_outcomes = 2   # Head or Tail
favorable_outcomes = 1  # Head

probability = favorable_outcomes / total_outcomes

print("Probability of getting Head:", probability)
```

## 2. Events and Outcomes

An event is a specific outcome or set of outcomes.

```python
# Example: Rolling a dice

dice = [1, 2, 3, 4, 5, 6]

# Event: Getting an even number
even_numbers = [2, 4, 6]

probability_even = len(even_numbers) / len(dice)

print("Probability of even number:", probability_even)
```

## 3. Conditional Probability

Conditional probability is the probability of an event given that another event has already occurred.

```python
# Example: Probability of drawing a red card from a deck (given it's a face card)

total_face_cards = 12   # J, Q, K of 4 suits
red_face_cards = 6      # Hearts + Diamonds

conditional_probability = red_face_cards / total_face_cards

print("Conditional Probability:", conditional_probability)
```

## 4. Random Experiments

Random experiments produce different outcomes each time.

```python
# Simulating coin toss using NumPy

coin_toss = np.random.choice(["Head", "Tail"], size=10)

print("Coin Toss Results:")
print(coin_toss)
```

## 5. Probability Simulation

We can simulate probability using large number of experiments.

```python
# Simulate 1000 coin tosses

tosses = np.random.choice(["Head", "Tail"], size=1000)

prob_head = np.sum(tosses == "Head") / 1000

print("Simulated Probability of Head:", prob_head)
```
