```python
import pandas as pd
```

## 1. Creating a Pandas Series

A **Series** is a one-dimensional labeled array capable of holding any data type.

```python
data = [10, 20, 30, 40, 50]

series = pd.Series(data)

series
```

## 2. Creating a DataFrame

A **DataFrame** is a two-dimensional table-like data structure consisting of rows and columns.

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [24, 27, 22, 32],
    "City": ["New York", "London", "Paris", "Tokyo"]
}

df = pd.DataFrame(data)

df
```

## 3. Viewing Data

Pandas provides several methods to quickly inspect the dataset.

```python
df.head()
```

```python
df.tail()
```

## 4. Dataset Information

We can get information about the dataset structure.

```python
df.info()
```

```python
df.describe()
```

## 5. Selecting Columns

We can select specific columns from a DataFrame.

```python
df["Name"]
```

```python
df[["Name", "Age"]]
```

## 6. Selecting Rows

Rows can be selected using the **iloc** method (integer position).

```python
df.iloc[0]
```

```python
df.iloc[1:3]
```

## 7. Filtering Data

We can filter data using conditional expressions.

```python
df[df["Age"] > 25]
```

## 8. Adding a New Column

```python
df["Salary"] = [50000, 60000, 45000, 70000]

df
```

## 9. Basic Data Analysis

```python
df["Age"].mean()
```
