```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

## 1. Creating a Sample Dataset

For demonstration purposes, we will create a small dataset containing missing values and duplicates.

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Alice"],
    "Age": [24, np.nan, 22, 32, 24],
    "City": ["New York", "London", "Paris", None, "New York"],
    "Salary": [50000, 60000, None, 70000, 50000]
}

df = pd.DataFrame(data)

df
```

## 2. Identifying Missing Values

We can check for missing values using the `isnull()` function.

```python
df.isnull()
```

```python
df.isnull().sum()
```

## 3. Handling Missing Values

There are multiple ways to handle missing values in a dataset.

### Removing Rows with Missing Values

```python
df_dropna = df.dropna()

df_dropna
```

### Filling Missing Values

```python
df_filled = df.fillna("Unknown")

df_filled
```

### Filling Missing Numeric Values with Mean

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)

df
```

## 4. Removing Duplicate Rows

Duplicate records can distort analysis results.

```python
df.duplicated()
```

```python
df_no_duplicates = df.drop_duplicates()

df_no_duplicates
```

## 5. Changing Data Types

Sometimes we need to convert columns to appropriate data types.

```python
df["Age"] = df["Age"].astype(int)

df.dtypes
```

## 6. Renaming Columns

```python
df.rename(columns={"Salary": "Annual_Salary"}, inplace=True)

df
```
