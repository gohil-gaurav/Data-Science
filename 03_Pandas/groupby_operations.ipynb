The `groupby()` function in Pandas allows us to group data based on specific columns and perform aggregation operations.

It is widely used in data analysis to summarize datasets.

---

## Topics Covered

1. Creating a Sample Dataset
2. Grouping Data
3. Aggregation Functions
4. Multiple Aggregations
5. Grouping by Multiple Columns

```python
import pandas as pd
```

## 1. Creating a Sample Dataset

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [60000, 50000, 65000, 70000, 52000, 62000],
    "Age": [25, 28, 24, 35, 30, 27]
}

df = pd.DataFrame(data)

df
```

## 2. Grouping Data

We can group data based on a column using the `groupby()` function.

```python
df.groupby("Department")
```

## 3. Mean Salary by Department

```python
df.groupby("Department")["Salary"].mean()
```

## 4. Total Salary by Department

```python
df.groupby("Department")["Salary"].sum()
```

## 5. Count Employees per Department

```python
df.groupby("Department")["Name"].count()
```

## 6. Multiple Aggregations

```python
df.groupby("Department")["Salary"].agg(["mean", "min", "max", "count"])
```

## 7. Grouping by Multiple Columns

```python
df.groupby(["Department", "Age"])["Salary"].mean()
```

## 8. Reset Index

```python
grouped_data = df.groupby("Department")["Salary"].mean().reset_index()

grouped_data
```

Key operations covered:
