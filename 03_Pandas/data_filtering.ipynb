Filtering is an important technique in data analysis that allows us to extract specific rows and columns based on conditions.

Using Pandas, we can easily filter datasets using logical conditions and built-in methods.

---

## Topics Covered

1. Creating a Sample Dataset
2. Filtering Rows
3. Filtering with Multiple Conditions
4. Using loc
5. Using iloc
6. Sorting Data

```python
import pandas as pd
```

## 1. Creating a Sample Dataset

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [24, 27, 22, 32, 29],
    "City": ["New York", "London", "Paris", "Tokyo", "Berlin"],
    "Salary": [50000, 60000, 45000, 70000, 65000]
}

df = pd.DataFrame(data)

df
```

## 2. Filtering Rows

We can filter rows based on conditions.

```python
df[df["Age"] > 25]
```

## 3. Filtering with Multiple Conditions

```python
df[(df["Age"] > 25) & (df["Salary"] > 60000)]
```

## 4. Using loc

The `loc` method is used to select rows and columns by labels.

```python
df.loc[df["Age"] > 25, ["Name", "Salary"]]
```

## 5. Using iloc

The `iloc` method selects rows and columns based on their index positions.

```python
df.iloc[0:3, 0:2]
```

## 6. Sorting Data

```python
df.sort_values(by="Salary")
```

```python
df.sort_values(by="Salary", ascending=False)
```

## 7. Filtering by Specific Values

```python
df[df["City"] == "London"]
```

## Conclusion

- Filtering rows based on conditions
- Filtering with multiple conditions
- Using loc and iloc

These operations are commonly used during data analysis and exploratory data analysis (EDA).
