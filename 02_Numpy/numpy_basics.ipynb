```python
import numpy as np
```

## 1. Creating NumPy Arrays

NumPy arrays are faster and more efficient than Python lists for numerical computations.

```python
arr1 = np.array([1, 2, 3, 4, 5])

print("1D Array:")
print(arr1)
```

```python
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("2D Array:")
print(arr2)
```

## 2. Array Shape and Dimensions

NumPy provides useful attributes to understand the structure of arrays.

```python
print("Shape of array:", arr2.shape)
print("Number of dimensions:", arr2.ndim)
print("Total elements:", arr2.size)
```

## 3. Creating Special Arrays

NumPy provides functions to generate arrays filled with zeros, ones, or a range of numbers.

```python
zeros_array = np.zeros((2,3))
print("Zeros Array:")
print(zeros_array)
```

```python
ones_array = np.ones((3,3))
print("Ones Array:")
print(ones_array)
```

```python
range_array = np.arange(0,10,2)

print("Range Array:")
print(range_array)
```

## 4. Array Operations

NumPy supports element-wise operations between arrays.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print("Addition:")
print(a + b)

print("Subtraction:")
print(a - b)

print("Multiplication:")
print(a * b)
```

## 5. Mathematical Functions

NumPy provides built-in mathematical functions for numerical computations.

```python
numbers = np.array([4,9,16])

print("Square Root:")
print(np.sqrt(numbers))

print("Mean:")
print(np.mean(numbers))

print("Sum:")
print(np.sum(numbers))
```

## 6. Array Indexing

Indexing allows us to access specific elements in an array.

```python
arr = np.array([10,20,30,40,50])

print("First element:", arr[0])
print("Last element:", arr[-1])
```

## 7. Array Slicing

Slicing allows us to extract portions of an array.

```python
print("Slice from index 1 to 4:")
print(arr[1:4])
```

## 8. Reshaping Arrays

Reshaping changes the structure of the array without changing its data.

```python
reshape_arr = np.arange(1,10)

reshaped = reshape_arr.reshape(3,3)

print("Reshaped Array:")
print(reshaped)
```

## 9. Random Numbers

NumPy can generate random numbers which are useful in simulations and machine learning.

```python
random_array = np.random.rand(3,3)

print("Random Array:")
print(random_array)
```

## 10. Aggregation Functions

Aggregation functions summarize the data in arrays.

```python
data = np.array([5,10,15,20])

print("Maximum value:", np.max(data))
print("Minimum value:", np.min(data))
print("Average value:", np.mean(data))
print("Total sum:", np.sum(data))
```
