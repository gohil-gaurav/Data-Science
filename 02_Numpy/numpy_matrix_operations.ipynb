```python
import numpy as np
```

## 1. Creating Matrices

In NumPy, matrices are represented as 2D arrays.

```python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)
```

## 2. Matrix Addition

Matrix addition is performed by adding corresponding elements of two matrices.

```python
result_add = A + B

print("Matrix Addition:")
print(result_add)
```

## 3. Matrix Subtraction

Matrix subtraction subtracts corresponding elements.

```python
result_sub = A - B

print("Matrix Subtraction:")
print(result_sub)
```

## 4. Matrix Multiplication

Matrix multiplication follows linear algebra rules.

NumPy provides multiple ways to perform this operation.

```python
result_mul = np.matmul(A, B)

print("Matrix Multiplication:")
print(result_mul)
```

## 5. Dot Product

The dot product is another way to multiply matrices.

```python
dot_product = np.dot(A, B)

print("Dot Product:")
print(dot_product)
```

## 6. Matrix Transpose

Transpose converts rows into columns and columns into rows.

```python
transpose_A = A.T

print("Transpose of Matrix A:")
print(transpose_A)
```

## 7. Identity Matrix

An identity matrix is a square matrix with ones on the diagonal and zeros elsewhere.

```python
identity_matrix = np.eye(3)

print("Identity Matrix:")
print(identity_matrix)
```

## 8. Matrix Inverse

The inverse of a matrix is a matrix that, when multiplied with the original matrix, results in the identity matrix.

```python
inverse_A = np.linalg.inv(A)

print("Inverse of Matrix A:")
print(inverse_A)
```

## Verification of Matrix Inverse

```python
verification = np.dot(A, inverse_A)

print("A × A⁻¹:")
print(verification)
```
