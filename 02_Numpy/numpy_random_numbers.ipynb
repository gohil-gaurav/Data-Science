Random number generation is widely used in data science, machine learning, simulations, and statistical modeling.

## Topics Covered

1. Random Integers
2. Random Floats
3. Random Arrays
4. Random Choice
5. Random Normal Distribution
6. Random Seed (Reproducibility)

```python
import numpy as np
```

## 1. Random Integer

NumPy can generate random integers within a specified range.

```python
random_int = np.random.randint(1, 100)

print("Random Integer:", random_int)
```

## 2. Random Float

Random floating numbers between 0 and 1 can be generated using NumPy.

```python
random_float = np.random.rand()

print("Random Float:", random_float)
```

## 3. Random Array

We can generate arrays filled with random values.

```python
random_array = np.random.rand(3,3)

print("Random 3x3 Array:")
print(random_array)
```

## 4. Random Choice

NumPy allows selecting random elements from an array.

```python
numbers = [10, 20, 30, 40, 50]

random_choice = np.random.choice(numbers)

print("Randomly selected number:", random_choice)
```

## 5. Random Normal Distribution

Normal distribution (also called Gaussian distribution) is widely used in statistics and machine learning.

```python
normal_distribution = np.random.randn(5)

print("Random numbers from normal distribution:")
print(normal_distribution)
```

## 6. Random Seed

Setting a random seed ensures reproducibility of results.  
This means the same random numbers will be generated every time the code runs.

```python
np.random.seed(42)

print("Random numbers with fixed seed:")
print(np.random.rand(3))
```

## Example: Simulating Random Data

```python
# Simulate 10 random values between 0 and 1

simulated_data = np.random.rand(10)

print("Simulated Data:")
print(simulated_data)
```
