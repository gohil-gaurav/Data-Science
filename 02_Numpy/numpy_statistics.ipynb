```python
import numpy as np
```

## 1. Creating Sample Data

We first create a sample NumPy array that we will use to demonstrate statistical operations.

```python
data = np.array([10, 20, 30, 40, 50])

data
```

## 2. Mean (Average)

The mean represents the average value of the data.

```python
mean_value = np.mean(data)

print("Mean:", mean_value)
```

## 3. Median

The median is the middle value of the dataset when the values are sorted.

```python
median_value = np.median(data)

print("Median:", median_value)
```

## 4. Standard Deviation

Standard deviation measures how spread out the numbers are in a dataset.

```python
std_value = np.std(data)

print("Standard Deviation:", std_value)
```

## 5. Variance

Variance measures the spread between numbers in the dataset.

```python
variance_value = np.var(data)

print("Variance:", variance_value)
```

## 6. Minimum and Maximum Values

NumPy provides functions to find the smallest and largest values in an array.

```python
print("Minimum value:", np.min(data))
print("Maximum value:", np.max(data))
```

## 7. Percentiles

Percentiles indicate the value below which a given percentage of observations fall.

```python
print("25th percentile:", np.percentile(data, 25))
print("50th percentile:", np.percentile(data, 50))
print("75th percentile:", np.percentile(data, 75))
```

## 8. Statistical Operations on 2D Arrays

```python
matrix = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

matrix
```

### Mean of entire matrix

```python
np.mean(matrix)
```

### Mean by rows

```python
np.mean(matrix, axis=1)
```

### Mean by columns

```python
np.mean(matrix, axis=0)
```
